**5. ¿Qué influencia tiene el transporte en la asistencia?**

```python
# Gráfico de boxplot para ver la relación entre transporte y asistencia
plt.figure(figsize=(5, 4))
sns.boxplot(x='TRANSPORTE', y='ASISTENCIA', data=df)
plt.title('Relación entre Transporte y Asistencia')
plt.xlabel('Tipo de Transporte')
plt.ylabel('Asistencia')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Calcular la media de la asistencia por tipo de transporte
media_asistencia_transporte = df.groupby('TRANSPORTE')['ASISTENCIA'].mean()
media_asistencia_transporte
```

El gráfico de boxplot muestra la relación entre el tipo de transporte utilizado por los estudiantes y su asistencia. Las medias de asistencia por tipo de transporte son las siguientes:

- **Mototaxi:**63.8
- **A pie:** 61.9
- **En Burro:** 49.0
-

**Análisis:** Los estudiantes que utilizan Mototaxi tienden a tener una asistencia ligeramente más alta en comparación con los que llegana a pie. Los que vienen en burro provienen de veredas y su asistencia es baja. Por tanto, se sugiere implementar una ruta de transporte escolar para estudiantes de vereda.
